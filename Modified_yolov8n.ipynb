```python
import os
import cv2
import torch
import shutil
import numpy as np
from pathlib import Path
from sklearn.model_selection import KFold
from ultralytics import YOLO
from ultralytics.data.utils import check_det_dataset
import albumentations as A
import matplotlib.pyplot as plt
from collections import Counter

# Class Names
class_names = ["EUR Pallets", "Non EUR Pallets"]

# Custom Modules 
class GhostModule(torch.nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=1, ratio=2, stride=1, relu=True):
        super().__init__()
        init_channels = int(out_channels / ratio)
        new_channels = out_channels - init_channels
        self.primary = torch.nn.Sequential(
            torch.nn.Conv2d(in_channels, init_channels, kernel_size, stride, kernel_size // 2, bias=False),
            torch.nn.BatchNorm2d(init_channels),
            torch.nn.ReLU(inplace=True) if relu else torch.nn.Identity(),
            torch.nn.Dropout(p=0.5)
        )
        self.cheap = torch.nn.Sequential(
            torch.nn.Conv2d(init_channels, new_channels, 3, 1, 1, groups=init_channels, bias=False),
            torch.nn.BatchNorm2d(new_channels),
            torch.nn.ReLU(inplace=True) if relu else torch.nn.Identity()
        )

    def forward(self, x):
        x1 = self.primary(x)
        x2 = self.cheap(x1)
        return torch.cat([x1, x2], dim=1)

class C3Ghost(torch.nn.Module):
    def __init__(self, in_channels, out_channels, n=1):
        super().__init__()
        self.blocks = torch.nn.Sequential(*[GhostModule(in_channels, out_channels) for _ in range(n)])
        self.shortcut = torch.nn.Identity()

    def forward(self, x):
        return self.shortcut(x) + self.blocks(x)

class BiODNeck(torch.nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.fuse = torch.nn.ModuleList([torch.nn.Conv2d(c, c, 3, padding=1) for c in channels])

    def forward(self, feats):
        return [f + conv(f) for f, conv in zip(feats, self.fuse)]

# Albumentations Augmentation
transform_base = A.Compose([
    A.RandomBrightnessContrast(p=0.5),
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.1),
    A.Affine(translate_percent=0.05, scale=(0.9, 1.1), rotate=(-10, 10), p=0.5),
    A.MotionBlur(p=0.2),
    A.RandomGamma(p=0.2),
    A.CLAHE(p=0.2),
    A.HueSaturationValue(p=0.2),
    A.GaussianBlur(p=0.1),
    A.RandomRotate90(p=0.3),
    A.CoarseDropout(max_holes=4, max_height=32, max_width=32, fill_value=0, p=0.2)
], bbox_params=A.BboxParams(format='yolo', label_fields=['class_labels'], min_visibility=0.3))

# Compute Class Weights 
def compute_class_weights(labels_dir, class_names):
    class_counts = Counter()
    total_labels = 0

    for label_file in Path(labels_dir).glob('*.txt'):
        with open(label_file, 'r') as f:
            for line in f:
                cls = int(line.strip().split()[0])
                class_counts[cls] += 1
                total_labels += 1

    class_weights = []
    for i in range(len(class_names)):
        count = class_counts.get(i, 0)
        weight = total_labels / (count + 1e-6)  # Avoid division by zero
        class_weights.append(weight)

    # Normalize weights to sum to number of classes
    norm_weights = [w * len(class_names) / sum(class_weights) for w in class_weights]

    print(f"Class counts: {class_counts}")
    print(f"Normalized class weights: {norm_weights}")
    return norm_weights

# Augment Dataset with Class Weights 
def augment_dataset(images_dir, labels_dir, output_img_dir, output_lbl_dir, augment_times=3, class_weights=None):
    output_img_dir.mkdir(parents=True, exist_ok=True)
    output_lbl_dir.mkdir(parents=True, exist_ok=True)

    for img_file in Path(images_dir).glob('*'):
        if img_file.suffix.lower() not in ['.jpg', '.jpeg', '.png']:
            continue
        img = cv2.imread(str(img_file))
        label_file = Path(labels_dir) / (img_file.stem + '.txt')
        if not label_file.exists():
            continue

        with open(label_file, 'r') as f:
            bboxes, class_labels = [], []
            for line in f:
                parts = line.strip().split()
                if len(parts) < 5:
                    continue
                try:
                    cls = int(float(parts[0]))
                    class_labels.append(cls)
                    bboxes.append(list(map(float, parts[1:])))
                except ValueError:
                    continue

        if class_weights:
            max_weight = max(class_weights[cls] for cls in class_labels)
            times = max(1, int(augment_times * max_weight))
        else:
            times = augment_times

        class_labels_names = [class_names[label] for label in class_labels]

        for i in range(times):
            augmented = transform_base(image=img, bboxes=bboxes, class_labels=class_labels_names)
            aug_img = augmented['image']
            # NO aspect ratio filtering here 
            aug_bboxes, aug_labels = augmented['bboxes'], augmented['class_labels']

            if not aug_bboxes:
                continue

            aug_img_name = f"{img_file.stem}_aug{i}.jpg"
            cv2.imwrite(str(output_img_dir / aug_img_name), aug_img)
            with open(output_lbl_dir / f"{img_file.stem}_aug{i}.txt", 'w') as f_out:
                for cls, bbox in zip(aug_labels, aug_bboxes):
                    cls_id = class_names.index(cls)
                    f_out.write(f"{cls_id} {' '.join(map(str, bbox))}\n")

#  Modify YOLOv8n 
def modify_yolov8(model_path):
    model = YOLO(model_path)
    for i, layer in enumerate(model.model.model):
        if isinstance(layer, torch.nn.Conv2d):
            model.model.model[i] = C3Ghost(layer.in_channels, layer.out_channels)
    model.model.neck = BiODNeck([128, 256, 512])
    return model

# Prepare Modified Model 
def prepare_modified_model(pretrained_path='yolov8n.pt', save_path='yolov8n_modified.pt'):
    model = modify_yolov8(pretrained_path)
    model.save(save_path)
    print(f"Modified model saved to {save_path}")
    return save_path

# K-Fold Training 
def run_k_fold(yaml_path, folds=3, modified_model_path='yolov8n_modified.pt',
               epochs=30, results_dir='runs_23_modified/crossval_results'):
    data = check_det_dataset(yaml_path)
    train_img_dir_original = Path(data['train'])
    train_lbl_dir_original = train_img_dir_original.parent / 'labels'

    # Compute class weights to address imbalance
    class_weights = compute_class_weights(train_lbl_dir_original, class_names)

    results_dir = Path(results_dir)
    results_dir.mkdir(parents=True, exist_ok=True)

    all_train_images = sorted([img for img in train_img_dir_original.iterdir()
                               if img.suffix.lower() in ['.jpg', '.jpeg', '.png']
                               and (train_lbl_dir_original / f"{img.stem}.txt").exists()])
    if not all_train_images:
        raise ValueError("No valid training images found!")

    kf = KFold(n_splits=folds, shuffle=True, random_state=42)
    results = []

    for fold, (train_idx, val_idx) in enumerate(kf.split(all_train_images)):
        print(f"\n--- Fold {fold+1}/{folds} ---")

        fold_dir = results_dir / f'fold{fold+1}'
        fold_train_img = fold_dir / 'train/images'
        fold_val_img = fold_dir / 'val/images'
        fold_train_lbl = fold_dir / 'train/labels'
        fold_val_lbl = fold_dir / 'val/labels'

        for d in [fold_train_img, fold_val_img, fold_train_lbl, fold_val_lbl]:
            d.mkdir(parents=True, exist_ok=True)

        def copy_files(indices, output_img_dir, output_lbl_dir):
            for idx in indices:
                img_path = all_train_images[idx]
                lbl_path = train_lbl_dir_original / f"{img_path.stem}.txt"
                shutil.copy(img_path, output_img_dir / img_path.name)
                shutil.copy(lbl_path, output_lbl_dir / lbl_path.name)

        copy_files(train_idx, fold_train_img, fold_train_lbl)
        copy_files(val_idx, fold_val_img, fold_val_lbl)

        # Augment training set with class-weighted times
        augment_dataset(fold_train_img, fold_train_lbl, fold_train_img, fold_train_lbl, augment_times=3, class_weights=class_weights)

        fold_yaml = fold_dir / 'data.yaml'
        with open(yaml_path, 'r') as src, open(fold_yaml, 'w') as dst:
            for line in src:
                if line.startswith('train:'):
                    dst.write(f"train: {fold_train_img.resolve()}\n")
                elif line.startswith('val:'):
                    dst.write(f"val: {fold_val_img.resolve()}\n")
                elif line.startswith('names:'):
                    dst.write(f"names: {class_names}\n")
                else:
                    dst.write(line)

        model = YOLO(modified_model_path)

        print(f"Training entire model for {epochs} epochs...")

        model.train(
            data=str(fold_yaml),
            epochs=epochs,
            imgsz=640,
            batch=16,
            lr0=0.00025,
            lrf=0.01,
            weight_decay=0.001,
            label_smoothing=0.1,
            patience=5,
            augment=True,
            mosaic=1.0,
            mixup=0.3,
            degrees=15.0,
            translate=0.1,
            scale=0.3,
            shear=2.0,
            perspective=0.0,
            project='/content/drive/MyDrive/trackmate/runs_23_modified',
            name=f'train_fold{fold+1}',
            exist_ok=True
        )

        best_model_path = Path(f'/content/drive/MyDrive/trackmate/runs_23_modified/train_fold{fold+1}/weights/best.pt')
        if best_model_path.exists():
            shutil.copy(best_model_path, fold_dir / 'best_model.pt')

        val_results = model.val()
        acc = val_results.box.map50
        print(f"Fold {fold+1} mAP50: {acc:.4f}")
        results.append(acc)

    summary_file = results_dir / 'results_summary.txt'
    with open(summary_file, 'w') as f:
        for i, r in enumerate(results):
            f.write(f"Fold {i+1} mAP@0.5: {r:.4f}\n")
        f.write(f"\nMean mAP@0.5: {np.mean(results):.4f} Â± {np.std(results):.4f}\n")

    plt.figure(figsize=(8, 4))
    plt.plot(range(1, folds + 1), results, marker='o', label='mAP50 per fold')
    plt.axhline(np.mean(results), color='r', linestyle='--', label='Mean mAP50')
    plt.xlabel("Fold")
    plt.ylabel("mAP@0.5")
    plt.title("Cross-Validation Performance")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    pretrained_path = 'yolov8n.pt'  # pretrained model path

    modified_model_path = prepare_modified_model(
        pretrained_path=pretrained_path,
        save_path='/content/drive/MyDrive/trackmate/yolov8n_modified.pt' # Modefied model path
    )

    yaml_file_path = '/content/drive/MyDrive/trackmate/dataset.yaml'
    run_k_fold(
        yaml_path=yaml_file_path,
        folds=3,
        modified_model_path=modified_model_path
    )
```
